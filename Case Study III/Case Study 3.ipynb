### Pivot with Id_b values in column

```python
#view data
df
```

```python
#create pivot table
df.pivot_table(values='ablePri2AxlesAuto', index='Id_a', columns = 'Id_b', fill_value=1)
```

```python
#creating pivot tables for all the columns and add another column 'pivoted_column'
pv1 = df.pivot_table(values='ablePri2AxlesAuto', index='Id_a', columns = 'Id_b', fill_value=-1)
pv1['pivoted_col'] = "ablePri2AxlesAuto"

pv2 = df.pivot_table(values='ablePri3AxlesAuto', index='Id_a', columns = 'Id_b', fill_value=-1)
pv2['pivoted_col'] = "ablePri3AxlesAuto"

pv3 = df.pivot_table(values='ablePri4AxlesAuto', index='Id_a', columns = 'Id_b', fill_value=-1)
pv3['pivoted_col'] = "ablePri4AxlesAuto"

pv4 = df.pivot_table(values='ablePri2AxlesDualTire', index='Id_a', columns = 'Id_b', fill_value=-1)
pv4['pivoted_col'] = "ablePri2AxlesDualTire"

pv5 = df.pivot_table(values='ablePri2AxlesTaxi', index='Id_a', columns = 'Id_b', fill_value=-1)
pv5['pivoted_col'] = "ablePri2AxlesTaxi"

pv6 = df.pivot_table(values='ablePri2AxlesTaxiPool', index='Id_a', columns = 'Id_b', fill_value=-1)
pv6['pivoted_col'] = "ablePri2AxlesTaxiPool"

pv7 = df.pivot_table(values='able23AxlesAuto', index='Id_a', columns = 'Id_b', fill_value=-1)
pv7['pivoted_col'] = "able23AxlesAuto"

pv8 = df.pivot_table(values='able24AxlesAuto', index='Id_a', columns = 'Id_b', fill_value=-1)
pv8['pivoted_col'] = "able24AxlesAuto"
```

```python
#merge all the pivot tables
df2 = pd.concat([pv1,pv2,pv3,pv4,pv5,pv6,pv7,pv8], axis=0)
```
